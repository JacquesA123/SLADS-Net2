# Nickel Concentration as a function of radius

```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Convert pixels to micrometers
def convert_pixels_to_micrometers(pixel_value, image, image_horizontal_field_width):
    return pixel_value * (image_horizontal_field_width / np.shape(image)[0]) * 1e6

image = np.zeros((1024, 1024))

# Example inputs
pixel_value = 99   # e.g. 50 pixels
field_width_meters = 6.24e-5  # 50 micrometers in meters
```

## Create master list of particles

```python
plotting_list = [] # This is the list of tuples to be plotted

# Create a list containing 4 empty lists (one for each i)
indexlists = [[] for _ in range(4)]
```

## Collect all the spectra folders into one list

```python
def find_spectra_folders(parent_folder):
    spectra_folders = []
    for root, dirs, files in os.walk(parent_folder):
        for d in dirs:
            if d.startswith("Spectra"):
                spectra_folders.append(os.path.join(root, d))
    return spectra_folders

# Path to your folder
folder_path = r"C:\PhenomData\NickelSpheres\Segmentation Project\Testing\Image Chain Tests\Test5\500_Particles"
spectra_folders = find_spectra_folders(folder_path)
```

## Given a Particle spectra folder, append all the spectra to the master list of particles

```python

for folder_path in spectra_folders:
    # Find all .npy files in the folder
    npy_files = glob.glob(os.path.join(folder_path, "*.npy"))

    # Sort files by modification time (oldest first)
    npy_files.sort(key=os.path.getmtime)

    # For each numpy file, compute the nickel peak counts, along with the total counts. Then find their ratio. This ratio is the nickel concentration
    for i, npy_file in enumerate(npy_files):
        if i > 3:
            break
        else:
            array = np.load(npy_file)
            total_counts = np.sum(array)
            # print(np.shape(array))
            # print(f'total counts = {total_counts}')
            nickel_peak_sum = np.sum(array[770:790])
            nickel_concentration = nickel_peak_sum / total_counts
            # print(f'nickel concentration = {nickel_concentration}')
            datapoint = (i, nickel_concentration)
            indexlists[i].append(nickel_concentration)
            plotting_list.append(datapoint)

    # print(tuples_list)
```

## Plot the nickel concentration as a function of radius

```python
x, y = zip(*plotting_list)

plt.plot(x, y, marker='o', linestyle='')  # dots only
plt.xlabel("First element of tuple")
plt.ylabel("Second element of tuple")
plt.title("Plot from List of Tuples")
plt.grid(True)
plt.show()
average0 = sum(indexlists[0]) / len(indexlists[0])
average1 = sum(indexlists[1]) / len(indexlists[1])
average2 = sum(indexlists[2]) / len(indexlists[2])
average3 = sum(indexlists[3]) / len(indexlists[3])
print(average0)
print(average1)
print(average2)
print(average3)
```

## Plot the nickel radial variation as a function of radius

```python
import os
import json
import numpy as np
import matplotlib.pyplot as plt

# -------- Settings --------
data_root = r"C:\PhenomData\NickelSpheres\Segmentation Project\Testing\Image Chain Tests\Test5\500_particles"  # top-level folder with your 200 image folders
bin_edges = np.arange(0, 11, 1)     # example: 4-5, 5-6, ..., 10 µm bins
element_index = 0                   # <-- index of Ni in your spectra array (adjust if needed)

# -------- Collect particle data --------
particle_radii = []
particle_stds = []

for image_folder in os.listdir(data_root):
    # Skip non-SEM image folders
    if not image_folder.startswith("SEM"):
        continue
    particles_path = os.path.join(data_root, image_folder, "Particles")
    # print(particles_path)
    if not os.path.isdir(particles_path):
        continue
    
    for particle_folder in os.listdir(particles_path):

        p_path = os.path.join(particles_path, particle_folder)
        if not os.path.isdir(p_path):
            continue
        
        
        # --- Read metadata to obtain radius ---
        metadata_path = os.path.join(p_path, "metadata.json")
        if not os.path.exists(metadata_path):
            continue
        with open(metadata_path, "r") as f:
            metadata = json.load(f)
        radius = metadata.get("radius", None)
        radius = convert_pixels_to_micrometers(radius, image, field_width_meters)
        # print(radius)
        if radius is None:
            continue

        # --- Load 4 spectra ---
        # print(p_path)
        spectra_path = os.path.join(p_path, f"Spectra_{particle_folder}")
        # print(spectra_path)
        # spectra = []
        concentrations = []
        for file in os.listdir(spectra_path):
            if file.endswith(".npy"):
                array = np.load(os.path.join(spectra_path, file))
                total_counts = np.sum(array)
                # spectra.append(array)
                nickel_peak_sum = np.sum(array[770:790])
                nickel_concentration = nickel_peak_sum / total_counts
                concentrations.append(nickel_concentration)
        
        nickel_standard_deviation = np.std(concentrations)
        
        
     
        

        particle_radii.append(radius)
        particle_stds.append(nickel_standard_deviation)

# -------- Bin and average --------
particle_radii = np.array(particle_radii)
particle_stds = np.array(particle_stds)

bin_means = []
for i in range(len(bin_edges)-1):
    low, high = bin_edges[i], bin_edges[i+1]
    mask = (particle_radii >= low) & (particle_radii < high)
    if np.any(mask):
        avg_std = np.mean(particle_stds[mask])
    else:
        avg_std = 0  # or np.nan if you prefer gaps
    bin_means.append(avg_std)

# -------- Plot --------
plt.bar(range(len(bin_means)), bin_means, width=0.8, tick_label=[f"{bin_edges[i]}-{bin_edges[i+1]}" for i in range(len(bin_means))])
plt.xlabel("Particle Radius (µm)")
plt.xlim(3.5, 8.5)
plt.ylabel("Average Standard Deviation of Nickel Concentration")
plt.title("Radial Variance in Nickel Concentration")
plt.show()
```
